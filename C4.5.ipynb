```python
from chefboost import Chefboost as chef
import pandas as pd
```

```python
df = pd.read_csv("C:/Users/raswa/Downloads/golf.txt")
df
```

```python
df['Decision']=df['Decision'].replace("Yes",1) 
```

```python
df['Decision']=df['Decision'].replace("No",0) 
```

```python
df.dtypes
```

```python
config = {'algorithm': 'C4.5'}
model = chef.fit(df, config = config, target_label = 'Decision')
```

```python
df
```

```python
prediction = chef.predict(model, param = ['Sunny', 'Hot', 'High', 'Weak'])
prediction
```

```python
prediction = chef.predict(model, param = ['Rain', 'Mild', 'Normal', 'Weak'])
prediction
```
